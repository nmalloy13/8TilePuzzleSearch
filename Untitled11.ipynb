```python
from collections import deque
#with open('input') as f:                                # reading in current state of puzzle
    #lines = f.readlines()
#start = lines[0].split(",")


class Node:

    def __init__(self, state, parent=None, action=None):
        self.state = state
        self.parent = parent
        self.action = action

    def __repr__(self):
        return self.action

def availableMoves(node):                          # returns all possible moves at a given state
    moves = []
    blank_index = node.state.index("_")
    if blank_index == 0:
        moves.append("Down")
        moves.append("Right")
    elif blank_index == 1:
        moves.append("Right")
        moves.append("Down")
        moves.append("Left")
    elif blank_index == 2:
        moves.append("Down")
        moves.append("Left")
    elif blank_index == 3:
        moves.append("Down")
        moves.append("Right")
        moves.append("Up")
    elif blank_index == 4:
        moves.append("Up")
        moves.append("Down")
        moves.append("Right")
        moves.append("Left")
    elif blank_index == 5:
        moves.append("Down")
        moves.append("Up")
        moves.append("Left")
    elif blank_index == 6:
        moves.append("Right")
        moves.append("Up")
    elif blank_index == 7:
        moves.append("Up")
        moves.append("Right")
        moves.append("Left")
    elif blank_index == 8:
        moves.append("Up")
        moves.append("Left")
    return moves

def addMoves(node):
    moves = availableMoves(node)
    children = []
    for move in moves:
        blankSpace = node.state.index("_")
        if move == "Left":
            tempNode = node.state.copy()
            tempNode[blankSpace], tempNode[blankSpace - 1] = tempNode[blankSpace - 1], tempNode[blankSpace]
            children.append([tempNode, tempNode[blankSpace] + "R"])
        elif move == "Right":
            tempNode = node.state.copy()
            tempNode[blankSpace], tempNode[blankSpace + 1] = tempNode[blankSpace + 1], tempNode[blankSpace]
            children.append([tempNode, tempNode[blankSpace] + "L"])
        elif move == "Up":
            tempNode = node.state.copy()
            tempNode[blankSpace], tempNode[blankSpace - 3] = tempNode[blankSpace - 3], tempNode[blankSpace]
            children.append([tempNode, tempNode[blankSpace] + "D"])
        elif move == "Down":
            tempNode = node.state.copy()
            tempNode[blankSpace], tempNode[blankSpace + 3] = tempNode[blankSpace + 3], tempNode[blankSpace]
            children.append([tempNode, tempNode[blankSpace] + "U"])
    return children
def bfs(start_state, goal_state):
    start_node = Node(start_state)
    goal_node = Node(goal_state)

    if start_state == goal_state:
        return ["Start state is the same as the goal state"]

    visited = set()
    queue = deque([start_node])

    while queue:
        current_node = queue.popleft()
        visited.add(tuple(current_node.state))

        if current_node.state == goal_state:
            path = []
            while current_node.parent:
                print(current_node.state)
                path.append(current_node.action)
                current_node = current_node.parent
            path.reverse()
            return path

        children = addMoves(current_node)
        for child_state, action in children:
            if tuple(child_state) not in visited:
                child_node = Node(child_state, parent=current_node, action=action)
                #print(len(child_node.action))
                queue.append(child_node)

    return ["No solution found"]


def heuristic(start,goal):
  h = 0
  for i in range(len(start)):
    if start[i]!=goal[i]:
      h+=1
  return h

def manhatten(start):
  dis = 0
  for i in range(1,9):
    if start[i]!="_":
        dis  += abs(i%3 - int(start[i])%3) + abs(i//3 - int(start[i])//3)
  return dis

def A_star(start,goal):
  open = []
  closed = set()
  start_node = Node(start)
  goal_node = Node(goal)
  open.append((manhatten(start),start_node))
  while open:
    open.sort(key=lambda x:x[0])
    _,current_node = open.pop(0)

    if current_node.state == goal_state:
      path = []
      while current_node:
        path.insert(0,current_node.action)
        current_node = current_node.parent
      return path

    closed.add(tuple(current_node.state))

    for cild,action in addMoves(current_node):
      if tuple(cild) not in closed:
        child_node = Node(cild,current_node,action)
        f_value = len(child_node.action) + manhatten(child_node.state)
        open.append((f_value,child_node))

  pass

def A_star_own(start,goal):
  open = []
  path = []
  closed = set()
  start_node = Node(start)
  closed.add(tuple(start_node.state))
  current_node = start_node
  while start_node.state!=goal:
    for child,action in addMoves(current_node):
      if tuple(child) not in closed:
        child_node = Node(child,current_node,action)
        f_value = len(child_node.action) + manhatten(child_node.state)
        open.append((f_value,child_node))
    open.sort(key = lambda x:x[0])
    _,current_node = open.pop(0)
    closed.add(tuple(current_node.state))
  while(current_node.parent!=None):
    path.append(0,current_node.action)
    current_node = current_node.parent
  return path

# Usage
start = ["1","4","2","3","_","5","6","7","8"]
goal_state = ["_","1","2","3","4","5","6","7","8"]
solution = A_star_own(start, goal_state)
if solution:
    print("Solution found with", len(solution), "moves:", solution)
else:
    print("No solution found")
```
